```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
```

# Scraping

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url_price = 'http://127.0.0.1:5000/api/v1.0/price'
browser.visit(url_price)
```

```python
# html = browser.html
# soup = bs(html, 'html.parser')
```

```python
price_df = pd.read_json(url_price)
price_df
```

```python
url_suburb = "http://127.0.0.1:5000/api/v1.0/suburb"
browser.visit(url_suburb)
# html = browser.html
# soup = bs(html, 'html.parser')
```

```python
cbd_df = pd.read_json(url_suburb)
cbd_df
```

```python
# Merge 

cbd_df.merge(price_df, how="inner", on=["Address"])
```

```python
browser.quit()
```
